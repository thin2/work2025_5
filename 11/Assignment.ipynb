```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the global temperature data from a URL
url = '2019.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
regionavg=data.groupby('Country or region')['Score'].mean().sort_values(ascending=False)
```

```python
regionavg.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(25, 6))
plt.bar(regionavg.index, regionavg.values)

plt.xticks(rotation=90)
plt.tight_layout()
plt.title("各地区幸福指数平均得分")

top5 = data.nlargest(5, "Score")
plt.figure()
plt.pie(top5["Score"], labels=top5["Country or region"], autopct="%1.1f%%")
plt.title("幸福指数最高的五个国家")
plt.show()
```
