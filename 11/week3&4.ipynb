```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

# 17
X_dice = np.array([1, 2, 3, 4, 5, 6])
P_dice = np.array([1/6] * 6)
CDF_dice = np.cumsum(P_dice)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(X_dice, P_dice, tick_label=X_dice)
plt.title("Probability Distribution Function of a Dice Roll")
plt.xlabel("Dice Outcome")
plt.ylabel("Probability")

plt.subplot(1, 2, 2)
plt.step(X_dice, CDF_dice, where='mid')
plt.title("Cumulative Distribution Function of a Dice Roll")
plt.xlabel("Dice Outcome")
plt.ylabel("Cumulative Probability")
plt.show()

# 18
X_coins = np.array([0, 1, 2, 3])
P_coins = np.array([1/8, 3/8, 3/8, 1/8])
CDF_coins = np.cumsum(P_coins)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(X_coins, P_coins, tick_label=X_coins)
plt.title("Probability Distribution Function of Coin Tosses")
plt.xlabel("Number of Tails")
plt.ylabel("Probability")

plt.subplot(1, 2, 2)
plt.step(X_coins, CDF_coins, where='mid')
plt.title("Cumulative Distribution Function of Coin Tosses")
plt.xlabel("Number of Tails")
plt.ylabel("Cumulative Probability")
plt.show()
# 19
sums = [sum(p) for p in product(range(1, 7), repeat=2)]
values, counts = np.unique(sums, return_counts=True)
P_dice_sum = counts / 36
CDF_dice_sum = np.cumsum(P_dice_sum)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(values, P_dice_sum, tick_label=values)
plt.title("Probability Distribution Function of Sum of Two Dice")
plt.xlabel("Sum")
plt.ylabel("Probability")

plt.subplot(1, 2, 2)
plt.step(values, CDF_dice_sum, where='mid')
plt.title("Cumulative Distribution Function of Sum of Two Dice")
plt.xlabel("Sum")
plt.ylabel("Cumulative Probability")
plt.show()

prob_sum_5_6_7_8 = np.sum(P_dice_sum[(values >= 5) & (values <= 8)])
print("P(X=5,6,7,8) =", prob_sum_5_6_7_8)

x_value = values[P_dice_sum == 1/12]
print("x such that P(X=x) = 1/12:", x_value)

queue_counts = np.array([27, 41, 21, 11])
queue_prob = queue_counts / 100
X_queue = np.array([0, 1, 2, 3])
CDF_queue = np.cumsum(queue_prob)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(X_queue, queue_prob, tick_label=X_queue)
plt.title("Probability Distribution Function of ATM Queue")
plt.xlabel("Queue Length")
plt.ylabel("Probability")

plt.subplot(1, 2, 2)
plt.step(X_queue, CDF_queue, where='mid')
plt.title("Cumulative Distribution Function of ATM Queue")
plt.xlabel("Queue Length")
plt.ylabel("Cumulative Probability")
plt.show()

queue_2_count = int(queue_prob[2] * 10)
queue_more_than_1 = int((queue_prob[2] + queue_prob[3]) * 10)
queue_less_than_3 = int((queue_prob[0] + queue_prob[1] + queue_prob[2]) * 10)

print("Expected number of times queue length is 2:", queue_2_count)
print("Expected number of times queue length is >1:", queue_more_than_1)
print("Expected number of times queue length is <3:", queue_less_than_3)
```
